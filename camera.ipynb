```python
# capturing img from web cam
cap = cv2.VideoCapture(0)
if cap.isOpened():
    ret,frame = cap.read()
    print(ret)
    print(frame)
else:
    ret = False

image = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title('My PiCTUYRE')
plt.axis('off')
plt.show()
cap.release()
```

```python
vdo = r"C:\Users\CTTC\Downloads\videoplayback (1).webm"

cv2.namedWindow(vdo)
cap = cv2.VideoCapture(vdo)
if cap.isOpened():
    ret,frame = cap.read()
    print(ret)
    print(frame)
else:
    ret = False

while ret:
    ret,frame = cap.read()
    if not ret:
        break
    cv2.imshow(vdo,frame)
    if cv2.waitKey(20)==27:
        break
cv2.destroyAllWindows()
cap.release()
```
